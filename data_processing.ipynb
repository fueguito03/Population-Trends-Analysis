```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Load the dataset
file_path = '인구동태건수_및_동태율_추이_출생_사망_혼인_이혼__.xlsx'
data = pd.read_excel(file_path, sheet_name='데이터')

# Rename columns for clarity
data.rename(columns={data.columns[0]: 'Category'}, inplace=True)

data.head(10)
```

```python
# Convert '2023' column to numeric and clean data
data['2023'] = pd.to_numeric(data['2023'], errors='coerce')
data.dropna(inplace=True)

# Save cleaned data
cleaned_file_path = 'population_trends_cleaned.csv'
data.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")
```
