```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

#출처: https://giveme-happyending.tistory.com/168 [소연의_개발일지:티스토리] // 한글폰트 깨짐코드
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

```

```python


# Load cleaned data
cleaned_file_path = './data/population_trends_cleaned.csv'
data = pd.read_csv(cleaned_file_path)

# Plot trends for selected categories
selected_categories = ['출생아수(명)', '사망자수(명)', '자연증가건수(명)']
eda_data = data[data['Category'].isin(selected_categories)].set_index('Category').transpose()

# Plot
for category in selected_categories:
    plt.plot(eda_data.index, eda_data.loc[:, category], label=category)

plt.title('Trends in Population Dynamics (2014-2023)')
plt.xlabel('Year')
plt.ylabel('Counts')
plt.legend()
plt.grid(True)
plt.show()
```
